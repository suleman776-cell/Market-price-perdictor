```python
df["OrderDate"]=pd.to_datetime(df["OrderDate"])
df["year"]=df["OrderDate"].dt.year
```

```python
df["Month"]=df["OrderDate"].dt.month    
df["Day"]=df["OrderDate"].dt.day
```

```python
df.drop(["OrderDate","OrderID"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
def fillnaobject(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mode()[0])

object_cols = ["Product", "Category", "Brand", "Platform", "City"]
fillnaobject(object_cols)
```

```python
def fillna_int_median(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].median())

int_cols = ["Day", "Quantity", "Month","year","Reviews"]
fillna_int_median(int_cols)
```

```python
# For Float / Integer Type
def fillnaFloatMean(cols):
    for i in cols:
        df[i] = df[i].fillna(df[i].mean())

numeric_cols = ["TotalAmount", "Price","Quantity"
    
]

fillnaFloatMean(numeric_cols)
```

```python
import  numpy as np
```

```python
from sklearn.preprocessing import LabelEncoder

def convert_object_to_int(cols):
    for col in cols:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])

objectcolumn = ["Product", "Category", "Brand", "Platform", "City"]
convert_object_to_int(objectcolumn)


```

```python
X=df.drop("Price",axis=1)
y=df["Price"]
```

```python
train_X, test_X, train_y, test_y = train_test_split(X,y, test_size=0.3, shuffle=False)
```

```python
print(train_X.shape,test_X.shape,train_y.shape,test_y.shape )
```

```python
from sklearn.ensemble import RandomForestRegressor

m = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

m.fit(train_X, train_y)
```

```python
a=m.predict(test_X)
a
```

```python
y_pred = m.predict(test_X)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

print("MAE:", mean_absolute_error(test_y, y_pred))
print("RMSE:", mean_squared_error(test_y, y_pred, squared=False))
print("R2 Score:", r2_score(test_y, y_pred))
```
